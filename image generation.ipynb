```python
images = ["lenna" , "airplane" , "mandril" , "pepper" , "sailboat" , "yacht"]
methods = ["VB-CIQ" , "FASTOCTREE" , "K-MEANS" , "MAXIMUM" , "MEDIAN CUT"]
```

```python
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
fig, ax = plt.subplots(nrows=1, ncols=5 , figsize=(100,100) )
for image in images:
    for i,col in enumerate(ax):
        im = Image.open(f"{methods[i]}/results/{image}/k8/reconstructed.png")
        col.imshow(im)

    fig.savefig(f"{image}ROW.png")
```
