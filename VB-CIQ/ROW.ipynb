```python
images = ["lenna" , "airplane" , "mandril" , "pepper" , "sailboat" , "yacht"]
```

```python
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
fig, ax = plt.subplots(nrows=6, ncols=6 , figsize = (20,20))
for i,image in enumerate(images):
    for j,col in enumerate(ax[i]):
        im = Image.open(f"results/{image}/k{2**(j+3)}/reconstructed.png")
        col.imshow(im)
        col.axis("off")

fig.savefig(f"ROW.png" , bbox_inches = "tight")
```
